# Interactive Analysis with Jupyter Notebook

This notebook explores a small sample dataset to understand user engagement and focus patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "user": ["Ava", "Ben", "Chloe", "Daniel", "Eva"],
    "sessions_per_week": [5, 3, 6, 2, 4],
    "focus_score": [80, 65, 90, 55, 75]
}

df = pd.DataFrame(data)
df
```

## Dataset Overview

The table above shows a simple dataset containing user names, weekly session counts, and focus scores.

```python
df.info()
df.describe()
```

```python
df.plot(x="user", y="focus_score", kind="line", marker="o")
plt.title("Focus Score by User")
plt.ylabel("Focus Score")
plt.show()
```

## Observation

Chloe has the highest focus score, while Daniel has the lowest. This may suggest that higher activity is linked to better focus, although a larger dataset would be needed to confirm that.
